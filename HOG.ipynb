```python
import cv2
import numpy as np
import os
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn import tree, svm
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from skimage.feature import hog
from skimage import feature


data = []
labels = []

count = 1

for filename in os.listdir("C:/Users/User/Downloads/archive (1)/casia/Train"):
    image = cv2.imread("C:/Users/User/Downloads/archive (1)/casia/Train/"+filename)
    gray = cv2.cvtColor( image, cv2.COLOR_RGB2GRAY )
    fd, hog_image = hog(gray, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualize=True)
    print("Image Loaded", count)
    count = count + 1
    i = 3072 - len(fd)
    for n in range(i):
        fd = np.append(fd, 0.0)
    
    labels.append(filename[:1])
    data.append(fd[:3072])
    
print("total images loaded", count)
```

```python
from sklearn.model_selection import train_test_split

train_data=[]
train_label=[]
test_data=[]
test_label=[]

train_data,test_data,train_label,test_label= train_test_split(data,labels,test_size=0.25,random_state=4)
```

```python
print(len(train_data),
len(train_label),
len(test_data),
len(test_label))
```

```python
logRegModel = LogisticRegression()
treegModel = tree.DecisionTreeClassifier(criterion='gini')
treeeModel = tree.DecisionTreeClassifier(criterion='entropy')
svmModel = svm.SVC()
naiveBayesModel = GaussianNB()
knnModel = KNeighborsClassifier()
RFModel = RandomForestClassifier()

logRegModel.fit(train_data, train_label)
treegModel.fit(train_data, train_label)
treeeModel.fit(train_data, train_label)
svmModel.fit(train_data, train_label)
naiveBayesModel.fit(train_data, train_label)
knnModel.fit(train_data, train_label)
RFModel.fit(train_data, train_label)


with open('HoglogRegModel.pickle', 'wb') as f1:
    pickle.dump(logRegModel, f1)
with open('HogtreegModel.pickle', 'wb') as f2:
    pickle.dump(treegModel, f2)
with open('HogtreeeModel.pickle', 'wb') as f3:
    pickle.dump(treeeModel, f3)
with open('HogsvmModel.pickle', 'wb') as f4:
    pickle.dump(svmModel, f4)
with open('HognaiveBayesModel.pickle', 'wb') as f5:
    pickle.dump(naiveBayesModel, f5)
with open('HogknnModel.pickle', 'wb') as f6:
    pickle.dump(knnModel, f6)
with open('HogRFModel.pickle', 'wb') as f7:
    pickle.dump(RFModel, f7)


print ("Classifiers have been saved")
```
